#### Oefenvragen voor deel 5: Tekstbestanden lezen en schrijven

Hieronder staat 1 oefenvraag die bij deel 5 van deze Python cursus hoort. Als je hierna nog meer wilt oefenen dan raden wij je aan om samen met je buurman of buurvrouw vragen te gaan maken die je vervolgens door de ander op laat lossen. 
<br/>


Vraag 1. Als oefening gaan we aan de slag met de temperatuur data uit de Bilt. Voor deze data gaan we 10-jarige gemiddeldes berekenen en die vervolgens met 2 decimalen gaat opslaan in een tekst bestand. Volg de volgende stappen:

Stap 1: Laad eerst weer het tekst bestand in met daarin de jaargemiddelde oppervlakte temperaturen uit de Bilt in voor de periode 1901 tot 1990 (Media/deBilt_yearly_temperatures.txt)

Stap 2: Maak een lege lijst waaraan je later de 10-jarige gemiddeldes kunt gaan toevoegen

Stap 3: Maak een for-loop waarin je in stappen van 10 door de hele lengte van de temperatuur data gaat (dus van 0 naar 80 in stappen van 10)

Stap 4: Bereken een 10-jarige gemiddelde en voeg de uitkomst toe aan de lijst die je in stap 2 gemaakt hebt.

Stap 5: Sla de 10-jarige gemiddeldes op in een nieuw tekstbestand en rond af op 2 decimalen.
