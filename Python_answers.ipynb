## Antwoorden deel 9 Debugging

1:

**Wat gaat er fout?** Er worden eindeloos veel getallen geprint. Het woord "einde" komt er nooit.

**Waarom gaat het fout?** Dit is een zogenaamde 'infinite-loop', een loop zonder einde. Zowel x als y worden iedere iteratie kleiner en zullen dus altijd kleiner blijven dan respectievelijk 20 en 10.

**Hoe krijg je het werkend?** Vervang "x -= 2" en/of "y -= 1" door bijvoorbeeld "x += 2" en/of door "y += 1".

<br/>

2:

**Wat gaat er fout?** Het antwoord is altijd nul

**Waarom gaat het fout?** Het 'return' statement ("return out") wordt gegeven direct na "out = 0" dus alles wat erna nog gebeurt met de variabele "out" wordt niet meer teruggegeven als uitkomst van de functie.        

**Hoe krijg je het werkend?** Door het 'return' statement te verplaatsen, bijvoorbeeld onderdeel van het 'if-statement'.



def divide(a,b):

    '''Function to divide two numbers'''
    
    out = 0
    
    if (a != 0 and b != 0):
    
        out = a/b

        return out
        
    else:
    
        print("One of the provided inputs is equal to zero")

**Waarom gaat het fout?** Omdat in Python, net als in de wiskune, 'delen door' voor 'optellen gaat. Er ontbreken dus haakjes.

**Hoe krijg je het werkend?** Vervang de regel "z = x+y/2" door "z = (x+y)/2"

**Hoe krijg je het werkend?** Vervang de regel "elif i > 10:" door "elif i >= 10:"

5:

**Wat gaat er fout?** De uitkomst is een som gelijk aan 2, terwijl de som gelijk zou moeten zijn aan 4

**Waarom gaat het fout?** Omdat de waarde van 'total' iedere iteratie weer op nul gezet wordt in plaats van enkel 1 keer voordat de loop begint.

**Hoe krijg je het werkend?** Door de regel "total = 0" voor de loop te plaatsen:

def som(numbers):

    #Bereken de som van de getallen in de reeks
    
    total = 0
    
    for number in numbers:
        
        total += number
        
    return total
    
<br/>

6:

**Wat gaat er fout?** Het aantal positieve getallen in deze reeks is 4, niet 5.

**Waarom gaat het fout?** Alle getallen kleiner dan nul worden gezien als negatief, maar alle andere getallen (ook het getal nul!) worden gezien als positief. Er is dus geen regel of uitzondering gemaakt voor het getal nul.

**Hoe krijg je het werkend?** Dat hangt er vanaf wat je wilt bereiken, maar als je het getal nul wilt overslaan, dan kan dat bijvoorbeeld door:

def isnegpos(numbers):

    #Test hoeveel waardes er positief zijn en hoeveel negatief
    
    neg = 0
    
    pos = 0
    
    for number in numbers:
    
        if number < 0:
        
            neg += 1
            
        elif number > 0:
        
            pos += 1
            
    return neg,pos
    
<br/>

**Wat gaat er fout?** Het lijkt op twee plekken fout te gaan, namelijk "Is 1 groter dan de voorgaande waarde -2? False" en "Is 2 groter dan de voorgaande waarde 0? False".

**Waarom gaat het fout?** In plaats van een waarde met de voorgaande waarde te vergelijken, word iedere waarde in deze code vergeleken met de laatste waarde uit de reeks ("numbers[-1]"). Die waarde is 2, waardoor het een aantal keer goed lijkt te gaan en een aantal keer fout. 

**Hoe krijg je het werkend?** Een werkende code zou er zo uit kunnen zien (er zijn ook andere oplossingen mogelijk!):

def islarger(numbers):

    #Tests of een getal in de reeks groter is dan de voorgaande waarde en geef True of False
    
    larger = ["nvt"]
    
    previous = numbers[0]
    
    for number in numbers[1:]:
    
        if number > previous:
        
            larger.append(True)
            
        else:
        
            larger.append(False)
            
        previous = number
        
    return larger
