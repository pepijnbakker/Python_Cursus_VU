Vraag 5

```python
def som(numbers):
    #Bereken de som van de getallen in de reeks
    for number in numbers:
        total = 0
        total += number
    return total

values = [1, -2, 1, -1, 3, 0, 2]
print("De som van de waardes is:", som(values))
```

Vraag 6

```python
def isnegpos(numbers):
    #Test hoeveel waardes er positief zijn en hoeveel negatief
    neg = 0
    pos = 0
    for number in numbers:
        if number < 0:
            neg += 1
        else:
            pos += 1
    return neg,pos

values = [1, -2, 1, -1, 3, 0, 2]
neg,pos = isnegpos(values)
print("Er zijn "+str(pos)+" positieve getallen")
print("Er zijn "+str(neg)+" negatieve getallen")
```

Vraag 7

```python
def islarger(numbers):
    #Tests of een getal in de reeks groter is dan de voorgaande waarde en geef True of False
    larger = ["nvt"]
    for number in numbers[1:]:
        if number > numbers[-1]:
            larger.append(True)
        else:
            larger.append(False)
    return larger

values = [1, -2, 1, -1, 3, 0, 2]
result = islarger(values)
print("Welke waardes zijn groter dan de voorgaande waarde:")
for i in range(1,len(values)):
    print("Is "+str(values[i])+" groter dan de voorgaande waarde "+str(values[i-1])+"? "+str(result[i]))
```
