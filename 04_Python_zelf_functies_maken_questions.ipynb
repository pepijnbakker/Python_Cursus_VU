#### Oefenvragen voor deel 4: Zelf functies maken

Hieronder een lijst oefenvragen behorende bij deel 4 van deze Python cursus. Als je hierna nog meer wilt oefenen dan raden wij je aan om samen met je buurman of buurvrouw vragen te gaan maken die je vervolgens door de ander op laat lossen. 

Vraag 1. Maak een functie die als input een lijst met getallen krijgt en als output de kleinste waarde terug geeft

Vraag 2. Verbeter de functie die je voor vraag 1 hebt gemaakt door in de functie eerst te controleren of de input wel van het type 'list' is.

Vraag 3. Maak een funcie waarmee je een lijst met getallen kan laten afronden op een x-aantal decimalen.
Tip: de 'round' functie werkt niet voor lijsten, maar je kunt met een loop werken waarin je de waardes in de lijst 1 voor 1 laat afronden.

Vraag 4. Maak een functie die zowel de omtrek als de inhoud van een bol kan berekenen voor een bepaalde radius afhankelijk van waar je om vraagt
